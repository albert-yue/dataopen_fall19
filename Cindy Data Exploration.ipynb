```python
col_list = list(imm_df)
col_list.remove('Year')
col_list.remove('Quarter')
col_list.remove('_quarter')
for col in col_list:
    res = imm_df[col].map(lambda s: re.sub(r'[a-zA-Z,]', '', str(s)))
    imm_df[col] = res.map(lambda s: '0' if len(s) == 0 else s)
imm_df = imm_df.astype({col: 'int64' for col in col_list})
```

```python
imm_df['Year'] = pd.to_datetime(imm_df['Year'],format="%Y").dt.year

def quarter_concatenation(row):
    year = str(row['Year'])
    if row['Quarter'] == "Q1":
        return_value = year + "-01-01"
    elif row['Quarter'] == "Q2":
        return_value = year + "-04-01"
    elif row['Quarter'] == "Q3":
        return_value = year + "-07-01"
    elif row['Quarter'] == "Q4":
        return_value = year + "-10-01"
    return return_value

imm_df['_quarter'] = pd.PeriodIndex(imm_df.apply(quarter_concatenation,axis=1), freq='Q').to_timestamp()
imm_df.head()
```
